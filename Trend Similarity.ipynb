```python
wcvecto=(worldcoin['AVG(PRICE)']).values
wcvector=wcvecto.reshape(-1,1)
print(wcvector)
```

```python
capture=[]

wcvector=(worldcoin['AVG(PRICE)']).values[:-1]



for coin in unique_coins:
    coin_data=df[df['crypto_name']==coin]
    coin_data.dropna( axis=0,  subset=None, inplace=False, ignore_index=False)
    if (coin_data['close'].size)<worldcoin['AVG(PRICE)'].size:
        continue
    for i in range (worldcoin['AVG(PRICE)'].size,coin_data['close'].size):
        df1=coin_data.iloc[i-worldcoin['AVG(PRICE)'].size:i]
        
        coinvector=(df1['close']).values[:-1]
        
        
        
        
        similarty=dot(wcvector,coinvector)/(norm(wcvector)*norm(coinvector))
        if(similarty>0.98):
            capture.append([coin,df1['date'].iloc[0]])

            
print(capture)
    
```
